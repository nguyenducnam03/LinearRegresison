```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
boston = load_boston()
datas = boston.data
labels = boston.target
```

```python
bos = pd.DataFrame(datas)
bos.columns = boston.feature_names
bos
```

```python
X_train,X_test,y_train,y_test = train_test_split(datas,labels,test_size = 0.2)
sklearn_reg = LinearRegression().fit(X_train,y_train)
print(sklearn_reg.coef_.shape)
print(X_test.shape)
print(y_test.shape)
print(y_test)
```

```python
y_predict = []
for i in range(X_test.shape[0]):
  cost = 0
  for j in range(sklearn_reg.coef_.shape[0]):
    cost += X_test[i][j]*sklearn_reg.coef_[j]
  y_predict.append(cost+sklearn_reg.intercept_)
# print(y_predict)
#error (cost function)
cost = 0
for i in range(len(y_predict)):
  cost += (y_predict[i] - y_test[i])**2
cost = math.sqrt(cost/len(y_predict))
print(cost)
```
